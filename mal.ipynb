```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
df.groupby(['sex','smoker'])[['total_bill']].agg('mean')
```

```python
df.pivot_table(index='sex',columns='smoker',values=['total_bill','tip'])
```

```python
df.pivot_table(index='sex',columns='smoker',values=['total_bill','tip'],aggfunc=['sum','std','count','median'])
```

```python
df.pivot_table(index=['sex','smoker'],columns=['day','time'],values=['total_bill'])
```
